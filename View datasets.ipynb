```python
import tensorflow as tf
import data_provider

from menpo.image import Image
```

```python
images, normals, masks = data_provider.MeinNormals256(1).get()
```

```python
sess = tf.Session()
```

```python
tf.train.start_queue_runners(sess=sess)
```

```python
im, nrm, mask = [Image.init_from_channels_at_back(x[0]) for x in sess.run([images, normals, masks])]
im.pixels += 111.
im.pixels /= 255.
im.pixels = im.pixels[[2, 1, 0]]
```
